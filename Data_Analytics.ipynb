```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

match = pd.read_csv("/content/matches.csv", encoding="latin-1")
delivery = pd.read_csv("/content/delivery.csv", encoding="latin-1")

teams_to_remove = ['Royal Challengers Bengaluru', 'Rising Pune Supergiants', 'Delhi Daredevils']

def plot_wins_after_removal(match, teams_to_remove):
    all_teams = match['team1'].tolist() + match['team2'].tolist()
    all_teams = list(set(all_teams))

    updated_teams = [team for team in all_teams if team not in teams_to_remove]

    match_filtered = match[~match['team1'].isin(teams_to_remove)]
    match_filtered = match_filtered[~match_filtered['team2'].isin(teams_to_remove)]
    match_filtered = match_filtered[~match_filtered['winner'].isin(teams_to_remove)]

    x = match_filtered['team1'].value_counts()
    y = match_filtered['team2'].value_counts()
    (x + y).plot(kind='barh')
    plt.title('Matches Played by Each Team (After Removal)')
    plt.xlabel('Number of Matches')
    plt.ylabel('Team Name')
    plt.show()

    x = pd.DataFrame({'winner': match_filtered['winner']}).value_counts()
    print(x)

    x.plot(kind='barh')
    plt.title('Number of matches won by each team (After Removal)')
    plt.xlabel('Number of wins')
    plt.ylabel('Team Name')
    plt.show()

plot_wins_after_removal(match, teams_to_remove)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

match = pd.read_csv("/content/matches.csv", encoding="latin-1")
delivery = pd.read_csv("/content/delivery.csv", encoding="latin-1")

def remove_teams(df, teams_to_remove):
    df = df[~df['team1'].isin(teams_to_remove)]
    df = df[~df['team2'].isin(teams_to_remove)]
    df = df[~df['winner'].isin(teams_to_remove)]
    return df

teams_to_remove = ['Royal Challengers Bengaluru', 'Rising Pune Supergiants', 'Delhi Daredevils']

filtered_match = remove_teams(match.copy(), teams_to_remove)

wins = filtered_match.pivot_table(index='team1', columns='team2', values='winner', aggfunc='count', fill_value=0)

total_matches = filtered_match.groupby('team1')['team1'].count() + filtered_match.groupby('team2')['team2'].count()

wins = wins.fillna(0)

plt.figure(figsize=(12, 8))
sns.heatmap(wins, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={'label': 'Number of Wins'})
plt.title('Heatmap of Wins Between Teams')
plt.xlabel('Opponent Team')
plt.ylabel('Team')
plt.show()
```
